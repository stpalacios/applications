### Constant Time O(1) - No matter the data, running time will always be constant

```python
#compare first two of a list
def first_compare(a,b):
    if a < b: 
        print (True)
    else: 
        print (False)

#start time -- timeit is a more accurate way to assess actual time but this is better for
#the demonstration purposes
start = time.time()
first_compare(data[0],data[1])
end = round(time.time() - start,4)

print ("Total Time:", end)


```

### Logarithmic Time — O(log n) - Reduces the size of the input data in each step  (e.g. binary tree, search)

```python
#Function to find the position of an item in a list
def binary_search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    
    while left <= right:
            middle = int((left + right) / 2) # cuts the list in half each time so reduces by log base 2
            if value < data[middle]:
                right = middle - 1
            elif value > data[middle]:
                left = middle + 1
            else:
                return middle
        
values = [9, 99, 9999, 99999, 999999]
for value in values:
    start = time.time()
    print("Value is located at: ", binary_search(data, value))
    end = round(time.time() - start,4)
    print (value , " took ", end , " seconds.\n")
```

### Linear Time — O(n) running time increases linearly--plots as a simple linear function  (A = xy +b) 

```python
#A linear search algorithm goes through each item in the list and sees if it matches input
def linear_search(data, value):
    for index in range(len(data)):
        if value == data[index]:
            return index


values = [9,99999,999999]
for value in values:
    start = time.time()
    print("Value is located at: ", linear_search(data, value))
    end = round(time.time() - start,4)
    print (value , " took ", end , " seconds.\n")
```

### Quasilinear Time — O(n log n) when each operation in the input data have a logarithm time complex

```python
#Code taken from MergeSort in Python at https://www.educative.io/edpresso/merge-sort-in-python (December 14, 2019)
def mergeSort(myList):
    if len(myList) > 1:
        mid = len(myList) // 2
        left = myList[:mid]
        right = myList[mid:]

        # Recursive call on each half
        mergeSort(left)
        mergeSort(right)

        # Two iterators for traversing the two halves
        i = 0
        j = 0
        
        # Iterator for the main list
        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
              # The value from the left half has been used
              myList[k] = left[i]
              # Move the iterator forward
              i += 1
            else:
                myList[k] = right[j]
                j += 1
            # Move to the next slot
            k += 1

        # For all the remaining values
        while i < len(left):
            myList[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            myList[k]=right[j]
            j += 1
            k += 1
            

values = [9, 99, 9999, 99999] #999999 --This would take several minutes
    
for value in values:
    my_list = random_list(value) #refers to the function at the beginning of this section 
    start = time.time()
    mergeSort(my_list)
    end = round(time.time() - start,4)
    print (value , " took ", end , " seconds.\n")
```
